```python
import pickle
import numpy as np
import igl
```

```python
with open('./staticdata/horlines.pkl', 'rb') as f:
    hl = pickle.load(f)
```

```python
len(hl), len(hl[0])
```

```python
theta = 0.067
cos_theta = np.cos(theta)
sin_theta = np.sin(theta)
rotation_matrix = np.array([[cos_theta, -sin_theta], [sin_theta, cos_theta]])
```

```python
ratio = 0.816993464052  # 1. 缩放比例 
# ratio = 0.8  # 1. 缩放比例 
```

```python
dist = (0, -0.15777788697145534, 0.025300056135217998)
```

```python
v2, t2 = igl.read_triangle_mesh('dataset/3DCaricShop/processedData/rawMesh/Donald Trump/C00001.obj')
```

```python
v2.shape
```

```python
v_list = []
for l_20 in hl:
    for a, b in l_20:
        v_list.append(b)
```

```python
v = np.array(v_list)
```

```python
v.shape
```

```python
v *= ratio
```

```python
for i, xyz in enumerate(v):
    v[i][1:] = np.dot(rotation_matrix, v[i][1:])
```

```python
v += dist
```

```python
c = 0
```

```python
hl
```

```python
new_hl = []
```

```python
for i, l_20 in enumerate(hl):
    new_l_20 = []
    for j, (a, b) in enumerate(l_20):
        new_l_20.append((a, v[c]))
        c += 1
    new_hl.append(new_l_20)
```
